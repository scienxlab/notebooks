# Custom colormaps

Let's look at a matplotlib colormap:

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
from collections import OrderedDict

viridis = cm.get_cmap('viridis', 12)
viridis
```

This is a `ListedColormap`. We can get an RGB color for a particular value &mdash; based on the range (0, 1):

```python
viridis(0.56)
```

Or look at all the colours:

```python
viridis.colors
```

## Modify viridis

```python
viridis = cm.get_cmap('viridis', 256)

newcolors = viridis(np.linspace(0, 1, 256))

pink = np.array([248/256, 24/256, 148/256, 1])
newcolors[:25, :] = pink
newcmp = ListedColormap(newcolors)

cms = [viridis, newcmp]

data = np.random.randn(30, 30)

fig, axs = plt.subplots(1, 2, figsize=(8, 3), constrained_layout=True)
for ax, cmap in zip(axs, cms):
    psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
    fig.colorbar(psm, ax=ax)
plt.show()
```

## Make a new ramp as `LinearSegmentedColormap`

```python
# Channel:  RED    GREEN     BLUE  ALPHA

red =   [100/256,   0/256,   0/256, 1.0]
white = [256/256, 256/256, 256/256, 1.0]
blue =  [  0/256,  50/256, 100/256, 1.0]

newcmp = LinearSegmentedColormap.from_list("foo", [red, white, blue])
```

```python
cms = [viridis, newcmp]

data = np.random.randn(30, 30)

fig, axs = plt.subplots(1, 2, figsize=(8, 3), constrained_layout=True)
for ax, cmap in zip(axs, cms):
    psm = ax.pcolormesh(data, cmap=cmap, rasterized=True, vmin=-4, vmax=4)
    fig.colorbar(psm, ax=ax)
plt.show()
```

Look up colours in a `LinearSegmentedColormap`...

```python
newcmp(0.5)
```

## Make a new ramp as `ListedColormap`

```python
# Channel:  RED    GREEN     BLUE  ALPHA

red =   [100/256,   0/256,   0/256, 1.0]
white = [256/256, 256/256, 256/256, 1.0]
blue =  [  0/256,  50/256, 100/256, 1.0]

newcmp = LinearSegmentedColormap.from_list("foo", [red, white, blue])

newlst = ListedColormap(newcmp(np.linspace(0, 1, 16)))
```

```python
newlst(0.5)
```

## Extract colours from an image

First, we'll need an image of the colourbar we want. It doesn't matter if it's horizontal or vertical (although you'll need to know which later.

```python
from PIL import Image

img = Image.open('data/arcgis_cmap_horizontal.png')
img
```

Now we can get this as a NumPy array.

```python
arr = np.asarray(img)[:, :, :3] / 255  # Remove alpha channel and rescale

plt.imshow(arr)
```

```python
arr.shape
```

Now we want to extract colours from here. It's 712 pixels long, and `matplotlib` colourmaps are only 256 pixels long. I'm not sure how `matplotlib` will deal with it but maybe the easiest thing to do might be to just take every 3rd pixel.

```python
colours = arr[0, ::3]  # First row, every third column.
```

```python
arccmap = LinearSegmentedColormap.from_list("arccmap", colours)
```

```python
data = np.random.uniform(0, 10, (20, 20))

fig, ax = plt.subplots(figsize=(4.6, 4), constrained_layout=True)
psm = ax.imshow(data, cmap=arccmap)
fig.colorbar(psm, ax=ax)
plt.show()
```

## Save as QGIS format

```python
XML = """<!DOCTYPE qgis_style>
<qgis_style version="0">
  <symbols/>
  <colorramps>
{0}
  </colorramps>
</qgis_style>"""

CRAMP = """    <colorramp type="gradient" name="{name}">
      <prop k="color1" v="{color1}"/>
      <prop k="color2" v="{color2}"/>
      <prop k="stops" v="{stops}"/>
    </colorramp>"""
```

```python
def get_rgba_from_cmap(cmap, value):
    """
    Look up from cmap object itself.
    """
    rgba = [str(int(round(255*v, 0))) for v in cmap(value)]
    return rgba
```

```python
def qgis_xml_from_cmap(cmap):
    params = {'name': cmap.name}
    params['color1'] = ','.join(get_rgba_from_cmap(cmap, 0.0))
    params['color2'] = ','.join(get_rgba_from_cmap(cmap, 1.0))
    stops = []
    for v in range(4, 100, 4):
        v /= 100
        s = f'{v};'
        s += ','.join(get_rgba_from_cmap(cmap, v))
        stops.append(s)
    params['stops'] = ':'.join(stops)
    return CRAMP.format(**params)
```

```python
with open(f"{arccmap.name}_as_QGIS.xml", 'wt') as f:
    f.write(XML.format(qgis_xml_from_cmap(arccmap)))
```
